#### The epilepsy dataset contains biomedical measurements of 195 individuals with and without the disease. The dataset has 23 columns, where the target column is "status" that indicates the presence (1) or absence (0) of the disease. The remaining columns are as follows:

   1. name: ASCII subject name and recording number
   2. MDVP:Fo(Hz): Average vocal fundamental frequency
   3. MDVP:Fhi(Hz): Maximum vocal fundamental frequency
   4. MDVP:Flo(Hz): Minimum vocal fundamental frequency
   5. MDVP:Jitter(%): MDVP jitter in percentage
   6. MDVP:Jitter(Abs): MDVP absolute jitter in ms
   7. MDVP:RAP: MDVP relative amplitude perturbation
   8. MDVP:PPQ: MDVP five-point period perturbation quotient
   9. Jitter:DDP: MDVP average absolute difference of differences between jitter cycles
   10. MDVP:Shimmer: MDVP local shimmer
   11. MDVP:Shimmer(dB): MDVP local shimmer in dB
   12. Shimmer:APQ3: MDVP three-point amplitude perturbation quotient
   13. Shimmer:APQ5: MDVP five-point amplitude perturbation quotient
   14. MDVP:APQ: MDVP overall amplitude perturbation quotient
   15. Shimmer:DDA: MDVP average absolute differences between the amplitudes of consecutive periods
   16. NHR: Noise to harmonics ratio
   17. HNR: Harmonics to noise ratio
   18. RPDE: Recurrence period density entropy measure
   19. DFA: Signal fractal scaling exponent
   20. spread1: Spectral spread
   21. spread2: D2 - Second Wiener-khinchin theorem
   22. D2: Correlation dimension
   23. PPE: Pitch period entropy

```python
## Here we will check the percentage of nan values present in each feature
## 1 -step make the list of features which has missing values
features_with_na=[features for features in df.columns if df[features].isnull().sum()>=1]
## 2- step print the feature name and the percentage of missing values

for feature in features_with_na:
    print(feature, np.round(df[feature].isnull().mean(), 4),  ' % missing values')

```

```python
df.columns
```

```python
## Replacing the Missing Values
dataset = df.copy()

for feature in features_with_na:
    ## We will replace by using median since there are outliers
    median_value=dataset[feature].median()
    
    ## create a new feature to capture nan values
    dataset[feature+'nan']=np.where(dataset[feature].isnull(),1,0)
    dataset[feature].fillna(median_value,inplace=True)
    
dataset[features_with_na].isnull().sum()
    
```

### Since there is a MDVP:PPQ and Jitter:DDP Two coloumns have missing values, we need to find the relationship between missing values and status

```python
# Group data by presence or absence of missing values in MDVP:PPQ column
mdvp_ppq_present = df[df['MDVP:PPQ'].notnull()]
mdvp_ppq_missing = df[df['MDVP:PPQ'].isnull()]

# Calculate mean value of status column for each group
mdvp_ppq_present_status_mean = mdvp_ppq_present['status'].mean()
mdvp_ppq_missing_status_mean = mdvp_ppq_missing['status'].mean()

# Print the mean values
print("Mean status value for MDVP:PPQ present:", mdvp_ppq_present_status_mean)
print("Mean status value for MDVP:PPQ missing:", mdvp_ppq_missing_status_mean)

# Group data by presence or absence of missing values in Jitter:DDP column
jitter_ddp_present = df[df['Jitter:DDP'].notnull()]
jitter_ddp_missing = df[df['Jitter:DDP'].isnull()]

# Calculate mean value of status column for each group
jitter_ddp_present_status_mean = jitter_ddp_present['status'].mean()
jitter_ddp_missing_status_mean = jitter_ddp_missing['status'].mean()

# Print the mean values
print("Mean status value for Jitter:DDP present:", jitter_ddp_present_status_mean)
print("Mean status value for Jitter:DDP missing:", jitter_ddp_missing_status_mean)
```

```python

# Create a copy of the dataframe
data = df.copy()

# Create a new column indicating whether MDVP:PPQ is missing or not
data['MDVP:PPQ_missing'] = data['MDVP:PPQ'].isnull()

# Create a new column indicating whether Jitter:DDP is missing or not
data['Jitter:DDP_missing'] = data['Jitter:DDP'].isnull()

# Create a scatter plot with different colors for missing and non-missing values of each feature
plt.scatter(data[data['MDVP:PPQ_missing']==False]['MDVP:PPQ'], 
            data[data['MDVP:PPQ_missing']==False]['Jitter:DDP'], 
            c=data[data['MDVP:PPQ_missing']==False]['status'], 
            cmap='viridis', label='MDVP:PPQ present')
plt.scatter(data[data['MDVP:PPQ_missing']==True]['MDVP:PPQ'], 
            data[data['MDVP:PPQ_missing']==True]['Jitter:DDP'], 
            c='red', label='MDVP:PPQ missing')
plt.scatter(data[data['Jitter:DDP_missing']==False]['MDVP:PPQ'], 
            data[data['Jitter:DDP_missing']==False]['Jitter:DDP'], 
            c=data[data['Jitter:DDP_missing']==False]['status'], 
            cmap='viridis', label='Jitter:DDP present')
plt.scatter(data[data['Jitter:DDP_missing']==True]['MDVP:PPQ'], 
            data[data['Jitter:DDP_missing']==True]['Jitter:DDP'], 
            c='orange', label='Jitter:DDP missing')

plt.xlabel('MDVP:PPQ')
plt.ylabel('Jitter:DDP')
plt.title('Relationship between MDVP:PPQ, Jitter:DDP, and Status')
plt.legend()
plt.show()
```

```python
# Create a copy of the data
data = df.copy()

# Add a new column indicating whether the value is missing or not
data['MDVP:PPQ_missing'] = data['MDVP:PPQ'].isnull().astype(int)
data['Jitter:DDP_missing'] = data['Jitter:DDP'].isnull().astype(int)

# Plot the relationship between missing values and status for each column
sns.barplot(x='MDVP:PPQ_missing', y='status', data=data)
plt.title('Relationship between missing values in MDVP:PPQ and status')
plt.show()

sns.barplot(x='Jitter:DDP_missing', y='status', data=data)
plt.title('Relationship between missing values in Jitter:DDP and status')
plt.show()
```

#### Here With the relation between the missing values and the dependent variable is clearly visible.So We need to replace these nan values with something meaningful


```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='mean')

# impute missing values for MDVP:PPQ
data['MDVP:PPQ'] = imputer.fit_transform(data[['MDVP:PPQ']])

# impute missing values for Jitter:DDP
data['Jitter:DDP'] = imputer.fit_transform(data[['Jitter:DDP']])
```

8. The 'PPE' (Pitch Period Entropy) feature has a wide range of values and a relatively high mean value, indicating that this feature may also be a useful predictor of Parkinson's disease.

```python
# create a list of numerical columns
num_cols = ['MDVP:Fo(Hz)', 'MDVP:Fhi(Hz)', 'MDVP:Flo(Hz)', 'MDVP:Jitter(%)', 'MDVP:Jitter(Abs)', 'MDVP:RAP', 
            'MDVP:PPQ', 'Jitter:DDP', 'MDVP:Shimmer', 'MDVP:Shimmer(dB)', 'Shimmer:APQ3', 'Shimmer:APQ5', 
            'MDVP:APQ', 'Shimmer:DDA', 'NHR', 'HNR', 'RPDE', 'DFA', 'spread1', 'spread2', 'D2', 'PPE']

# create subplots
fig, axs = plt.subplots(ncols=4, nrows=6, figsize=(20,20))
axs = axs.flatten()

# plot the distribution of each numerical feature with respect to the target variable
for i, col in enumerate(num_cols):
    sns.histplot(data=df, x=col, hue='status', kde=True, ax=axs[i])
    axs[i].set_title(col)

plt.tight_layout()
plt.show()
```

#### CHECKING OUTLIERS

```python
fig, axs = plt.subplots(ncols=4, nrows=6, figsize=(20,20))
axs = axs.flatten()

for i, col in enumerate(num_cols):
    sns.boxplot(x='status', y=col, data=df, ax=axs[i])

plt.tight_layout()
plt.show()
```

##### The above code generates a grid of boxplots, where each plot represents the distribution of the numerical column for each value of the status variable (0 or 1). The box in the boxplot represents the interquartile range (IQR), and the whiskers represent the minimum and maximum values, excluding outliers. Outliers are shown as points beyond the whiskers.

##### By examining these boxplots, we can see if there are any significant differences in the distribution of each numerical variable between the two categories of the status variable.
